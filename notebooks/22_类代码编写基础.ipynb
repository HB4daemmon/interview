## 知识点 1：class 与 __init__：实例化与初始化

__init__ 用于初始化实例属性；self 指向当前实例。


```python
class User:
    def __init__(self, name):
        self.name = name

u = User('Ada')
print(u.name)
```

## 知识点 2：实例属性 vs 类属性：共享与独立

类属性属于类本身，所有实例共享；实例属性属于实例自己。


```python
class User:
    role = 'member'  # 类属性
    def __init__(self, name):
        self.name = name

u1 = User('Ada')
u2 = User('Bob')
print(u1.role, u2.role, User.role)
User.role = 'admin'
print(u1.role, u2.role)
```

## 知识点 3：self：方法调用的隐式第一个参数

u.method(x) 实际等价于 Class.method(u, x)。


```python
class A:
    def f(self, x):
        return (self, x)

a = A()
print(a.f(1))
print(A.f(a, 1))
```

## 知识点 4：@classmethod：替代构造器/工厂方法

classmethod 的第一个参数是 cls（类），常用于提供多种构造方式。


```python
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y

    @classmethod
    def from_tuple(cls, t):
        x, y = t
        return cls(x, y)

p = Point.from_tuple((3, 4))
print(p.x, p.y)
```

## 知识点 5：@staticmethod：放在类命名空间里的工具函数

staticmethod 不自动传 self/cls，通常用于与类概念相关的工具函数。


```python
class Math:
    @staticmethod
    def add(a, b):
        return a + b

print(Math.add(1, 2))
```

## 知识点 6：属性查找：实例 -> 类 -> 父类

如果实例上没有该属性，会到类上找；再到父类链上找。


```python
class Base:
    x = 1

class Child(Base):
    pass

c = Child()
print(c.x)
Child.x = 2
print(c.x)
```

## 综合小案例：实现 Rectangle：面积与替代构造器

写 Rectangle(w,h) 提供 area；再写 from_square(side) 类方法。


```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    @classmethod
    def from_square(cls, side):
        return cls(side, side)

r = Rectangle(3, 4)
print(r.area())
print(Rectangle.from_square(5).area())
```
