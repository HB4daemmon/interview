## 知识点 1：def/return：函数定义与返回值

return 会结束函数并返回值；没有 return 时默认返回 None。


```python
def greet(name):
    return f'Hello, {name}!'

print(greet('Ada'))

def f():
    pass

print(f())  # None
```

## 知识点 2：参数与局部变量：函数内部是局部作用域

函数内赋值的变量默认是局部变量；作用域细节会在第 14 节展开。


```python
def add(a, b):
    c = a + b
    return c

print(add(1, 2))
```

## 知识点 3：函数是一等对象：可以当作参数传入

把函数当作值传递，能实现更通用的“高阶函数”。


```python
def apply(fn, x, y):
    return fn(x, y)

def mul(x, y):
    return x * y

print(apply(mul, 2, 3))
```

## 知识点 4：返回函数：函数工厂与简单闭包（预告）

函数可以返回函数；返回的函数可以携带外层状态（闭包）。


```python
def make_multiplier(k):
    def mul(x):
        return x * k
    return mul

mul2 = make_multiplier(2)
print(mul2(10))
```

## 知识点 5：递归：一定要有终止条件

递归适合分治问题；注意终止条件与递归深度限制。


```python
def factorial(n):
    if n < 0:
        raise ValueError('n must be >= 0')
    if n in (0, 1):
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

## 知识点 6：简单自测：assert

assert 适合学习/调试阶段的快速验证；生产中通常用测试框架。


```python
def is_even(n):
    return n % 2 == 0

assert is_even(2)
assert not is_even(3)
print('ok')
```

## 常见坑

- 递归深度有限（过深可能 RecursionError）


## 综合小案例：实现 is_palindrome：回文判断（忽略大小写与空格）

把输入规范化后再判断：只保留字母数字，统一小写，然后比较正序与逆序。


```python
import re

def is_palindrome(s: str) -> bool:
    t = re.sub(r'[^0-9a-zA-Z]+', '', s).lower()
    return t == t[::-1]

print(is_palindrome('A man, a plan, a canal: Panama'))
print(is_palindrome('hello'))
```
